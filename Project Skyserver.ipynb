#### Adding new variables

Two new variables will be added. The variable *velocity* which will be numeric, and *super_vel* which will be binary. The *super_vel* will indicate if the celestial object has a velocity greater than the mean velocity of the dataset.

To calculate the velocity of the celestial object we will use the equation:

$$
v = c · z
$$

Where c is the speed of light in *Km/s* and z is the redshift

```python
c = 300000
sdss['velocity'] = sdss['redshift']*c
```

```python
sdss['velocity'].describe()
```

```python
sdss['super_vel'] = ""

sdss['super_vel'][sdss['velocity'] >= np.mean(sdss['velocity'])] = 1
sdss['super_vel'][sdss['velocity'] <  np.mean(sdss['velocity'])] = 0
```

```python
sdss['super_vel'].value_counts()
```

We discretize the redshift and the velocity for the sake of visualization.

```python
sdss['disc_redshift'] = pd.cut(sdss['redshift'], bins = range(-1,9,1), labels =["-1","0","1","2","3","4","5","6","7"])
sdss['disc_vel'] = pd.cut(sdss['velocity'], bins = [-10000,0,10000,20000,30000,40000,50000,60000,70000,80000,90000,100000, 2200000], labels = ["-10000","0","10000","20000","30000","40000","50000","60000","70000","80000","90000","100000"])
```

# K Nearest Neighbors

First of all we scale the data frame, dropping the column we want to predict.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(sdss.drop('class',axis=1))
```

```python
scaled_features = scaler.transform(sdss.drop('class',axis=1))
```

```python
sdss_feat = pd.DataFrame(scaled_features, columns=['ra','dec','u','g','r','i','z','redshift','velocity','super_vel','disc_redshift','disc_vel'])
sdss_feat.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = sdss_feat
y = sdss['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

## Predictions and Evaluation

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test, pred))
```

### Choosing a K Value

```python
error_rate = []
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,10), error_rate, color = 'blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.xlabel('K value')
plt.ylabel('Error Rate')
plt.title('Error Rate vs K Value')
```

We see that the optimal K value is 3

### Retrain with K = 3

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
